```python
import pandas as pd
from utils import load_json
from mlxtend.preprocessing import TransactionEncoder

champion_builds = load_json('outputs/champion_item_lists.json')
champion_builds
```

```python
te = TransactionEncoder()
champion_builds_df = {}
for champion, builds in champion_builds.items():
    te_ary = te.fit_transform(builds)
    df = pd.DataFrame(te_ary, columns=te.columns_)
    champion_builds_df[champion] = df
champion_builds_df
```

```python
from mlxtend.frequent_patterns import apriori, fpgrowth, fpmax

def get_frequent_builds_with_algorithm(df, algorithm, min_support):
    champion_frequent_builds = {}
    for champion, builds in df.items():
        frequent = algorithm(builds, min_support=min_support, use_colnames=True)
        champion_frequent_builds[champion] = frequent
    return champion_frequent_builds
```

```python
frequent_apriori = get_frequent_builds_with_algorithm(champion_builds_df, apriori, 0.4)
for champion, builds in frequent_apriori.items():
    print(f'Champion {champion}')
    display(champion_builds[champion])
    display(builds)
```

```python
frequent_fpgrowth = get_frequent_builds_with_algorithm(champion_builds_df, fpgrowth, 0.4)
for champion, builds in frequent_fpgrowth.items():
    print(f'Champion {champion}')
    display(champion_builds[champion])
    display(builds)
```

```python
frequent_fpmax = get_frequent_builds_with_algorithm(champion_builds_df, fpmax, 0.4)
for champion, builds in frequent_fpmax.items():
    print(f'Champion {champion}')
    display(champion_builds[champion])
    display(builds)
```

```python
from pymongo import MongoClient
from bson.codec_options import CodecOptions, TypeEncoder, TypeRegistry

DF_TO_JSON_TYPE = 'records'

class FrozensetEncoder(TypeEncoder):
    python_type = frozenset
    def transform_python(self, value):
        return list(value)

type_registry = TypeRegistry([FrozensetEncoder()])
codec = CodecOptions(type_registry=type_registry)
```

```python
client = MongoClient()
db = client.better_builds
collection = db.get_collection('frequent_builds', codec_options=codec)
collection.drop()

insert_builds = []

for champion, builds in frequent_fpmax.items():
    builds_dict = builds.to_dict(orient=DF_TO_JSON_TYPE)
    insert_build = {
        'type': 'basic',
        'champion': int(champion),
        'builds': builds_dict
    }
    insert_builds.append(insert_build)
    
for champion, builds in frequent_fpgrowth.items():
    builds_dict = builds.to_dict(orient=DF_TO_JSON_TYPE)
    insert_build = {
        'type': 'advanced',
        'champion': int(champion),
        'builds': builds_dict
    }
    insert_builds.append(insert_build)

collection.insert_many(insert_builds)
```
